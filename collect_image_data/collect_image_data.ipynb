```python
# import mydownloader


# Function to download images for a given course
def download_images(course_name, num_images=2):
    # Set the directory and file extension
    my_downloader.directory = 'images'
    my_downloader.extensions = '.jpg'
    
    # Download images
    my_downloader.download(course_name, limit=num_images, verbose=True)
    
    # Rename and move downloaded images
    downloaded_files = os.listdir('images/' + course_name)
    for i, file in enumerate(downloaded_files):
        new_file_name = f"{course_name}_{i+1}.jpg"
        os.rename(f'images/{course_name}/{file}', f'images/{new_file_name}')
    
    # Remove the empty directory
    os.rmdir('images/' + course_name)
```

```python
from PIL import Image
from io import BytesIO
import time
# Function to download images for a given course until images are loaded
def download_images(course_name, num_images=2):
    search_url = f"https://www.google.com/search?hl=en&tbm=isch&q={course_name}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
    }
    
    response = requests.get(search_url, headers=headers)
    time.sleep(5)  # Wait for 5 seconds to allow images to load
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    images = soup.find_all('img')[1:]  # Skip the first image which is usually the Google logo
    
    downloaded_images = 0
    for i, img in enumerate(images):
        if downloaded_images >= num_images:
            break
        try:
            img_url = img['src']
            img_data = requests.get(img_url).content
            with open(f'images/{course_name}_{downloaded_images + 1}.jpg', 'wb') as handler:
                handler.write(img_data)
            downloaded_images += 1
        except Exception as e:
            print(f"Error downloading image: {e}")

# Loop through each course in the DataFrame and download images
for course in data_course['Mata Kuliah']:
    download_images(course)

print("Images downloaded successfully.")
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Function to download images for a given course
def download_images(course_name, num_images=2):
    # Initialize the Chrome driver
    driver = webdriver.Chrome()
    
    try:
        # Open Google Images and search for the course
        driver.get("https://www.google.com/imghp")
        search_box = driver.find_element_by_css_selector("input[name='q']")
        search_box.send_keys(course_name)
        search_box.send_keys(Keys.RETURN)
        
        # Wait for the search results to load
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "img.Q4LuWd"))
        )
        
        # Click on the first image to open it
        first_image = driver.find_element(By.CSS_SELECTOR, "img.Q4LuWd")
        first_image.click()
        
        # Wait for the image to load fully
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "img.n3VNCb"))
        )
        
        # Get the URL of the opened image
        image_url = driver.find_element(By.CSS_SELECTOR, "img.n3VNCb").get_attribute("src")
        
        # Download the image
        response = requests.get(image_url)
        with open(f'images/{course_name}_1.jpg', 'wb') as handler:
            handler.write(response.content)
        
    finally:
        # Close the browser
        driver.quit()

# Loop through each course in the DataFrame and download images
for course in data_course['Mata Kuliah']:
    download_images(course)

print("Images downloaded successfully.")
```
