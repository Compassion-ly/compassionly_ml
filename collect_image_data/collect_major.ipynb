```python
import pandas as pd

data = pd.read_csv('../dataset/finished_major_data.csv')
data
```

```python
major_data = data['NAMA'].tolist()
major_string = ','.join(major_data)
major_string
```

```python
file_path = 'major_list.txt'

# Open the file in write mode and write the string to it
with open(file_path, 'w') as file:
    file.write(major_string)

print("String saved to major_list.txt")
```

```python
univ_data = pd.read_csv('../dataset/univ_data.csv')
univ_data
```

```python
# get unique data 'PERGURUAN TINGGI'
data_univ = univ_data['PERGURUAN TINGGI'].unique()
data_univ
```
