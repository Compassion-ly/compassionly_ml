```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from transformers import BertTokenizer, TFBertModel
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python
data = pd.read_csv('major_data_cleaned_300.csv')
data['combined_features'] = data['Description'] + ' ' + data['Mata Kuliah']
label_encoder = LabelEncoder()
data['encoded_labels'] = label_encoder.fit_transform(data['NAMA'])
```

```python
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data['combined_features'], data['encoded_labels'], test_size=0.2, random_state=42)
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
train_encodings = tokenizer(X_train.tolist(), truncation=True, padding=True, max_length=128)
test_encodings = tokenizer(X_test.tolist(), truncation=True, padding=True, max_length=128)

```

```python
import tensorflow as tf
def create_dataset(encodings, labels):
    return tf.data.Dataset.from_tensor_slices((
        dict(encodings),
        labels
    ))
```

```python
train_dataset = create_dataset(train_encodings, y_train)
test_dataset = create_dataset(test_encodings, y_test)
```

```python
train_dataset
```

```python
bert_model = TFBertModel.from_pretrained('bert-base-multilingual-cased')

input_ids = Input(shape=(1, ), dtype=tf.int32, name='input_ids')
attention_mask = Input(shape=(128,), dtype=tf.int32, name='attention_mask')

bert_outputs = bert_model(input_ids, attention_mask=attention_mask)
pooled_output = bert_outputs[1]

dropout = Dropout(0.3)(pooled_output)
output = Dense(len(label_encoder.classes_), activation='softmax')(dropout)

model = Model(inputs=[input_ids, attention_mask], outputs=output)
model.compile(optimizer=Adam(learning_rate=2e-5), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
tf.config.run_functions_eagerly(True)
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

model.fit(
    train_dataset.shuffle(1000).batch(8),
    epochs=10,
    validation_data=test_dataset.batch(8),
    callbacks=[early_stopping]
)
```

```python
def recommend_major(input_text):
    input_encodings = tokenizer([input_text], truncation=True, padding=True, max_length=128)
    input_ids = tf.constant(input_encodings['input_ids'])
    attention_mask = tf.constant(input_encodings['attention_mask'])

    predictions = model.predict([input_ids, attention_mask])
    top_10_indices = np.argsort(predictions[0])[-10:][::-1]
    return label_encoder.inverse_transform(top_10_indices)

# Example usage
input_text = "Saya ingin menjadi seorang ahli teknologi yang bisa bisnis"
recommended_majors = recommend_major(input_text)
print("Top-10 recommended majors:", recommended_majors)
```
