```python
# def split_mata_kuliah(mata_kuliah):
#     mata_kuliah_list = mata_kuliah.split('\n')
#     mata_kuliah_list = [mk.split('. ', 1)[1] if '. ' in mk else mk for mk in mata_kuliah_list]
#     return mata_kuliah_list

# only takes 5 courses
def split_mata_kuliah(mata_kuliah):
    mata_kuliah_list = mata_kuliah.split('\n')
    mata_kuliah_list = mata_kuliah_list[:5]
    mata_kuliah_list = [mk.split('. ', 1)[1] if '. ' in mk else mk for mk in mata_kuliah_list]
    return mata_kuliah_list
```

```python
data['Split Mata Kuliah'] = data['Mata Kuliah'].apply(split_mata_kuliah)
# data['Split Mata Kuliah'] = data['Split Mata Kuliah'][:5]

# data_exploded = data.explode('Split Mata Kuliah').reset_index(drop=True)
# data_exploded = data.explode('Split Mata Kuliah').reset_index(drop=True)
```

```python
data_courses = data_exploded[['Split Mata Kuliah', 'NAMA', 'Matematika', 'Sains', 'Fisika', 'Sosiologi', 'Biologi', 'Kimia', 'Bisnis dan Ekonomi', 'Teknologi', 'Seni', 'Sastra dan Linguistik', 'Pendidikan', 'Hukum', 'Lingkungan', 'Kesehatan', 'Geografi', 'Komunikasi', 'Sejarah dan Filsafat', 'Hukum']]
data_courses
```

```python
data_courses
```

```python
data_courses.drop_duplicates(subset='Split Mata Kuliah', inplace=True)
data_courses.reset_index(drop=True, inplace=True)
data_courses
```

```python
data_courses.info()
```

```python
columns_to_consider = ['Matematika', 'Sains', 'Fisika', 'Sosiologi', 'Biologi', 'Kimia',
                       'Bisnis dan Ekonomi', 'Teknologi', 'Seni', 'Sastra dan Linguistik',
                       'Pendidikan', 'Hukum', 'Lingkungan', 'Kesehatan', 'Geografi',
                       'Komunikasi', 'Sejarah dan Filsafat']

index_mapping = {column: index+1 for index, column in enumerate(columns_to_consider)}

# Find the column with the maximum value for each row
max_columns = data_courses[columns_to_consider].idxmax(axis=1)

# Map the column names to their respective index values
data_courses['highest'] = max_columns.map(index_mapping)
```
