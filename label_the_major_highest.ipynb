```python
import pandas as pd

data = pd.read_csv('data_major_with_information.csv')
data_validations = pd.read_csv('major_data_validation.csv')
```

```python
data
```

```python
columns_to_consider = ['Matematika', 'Sains', 'Fisika', 'Sosiologi', 'Biologi', 'Kimia',
                       'Bisnis dan Ekonomi', 'Teknologi', 'Seni', 'Sastra dan Linguistik',
                       'Pendidikan', 'Hukum', 'Lingkungan', 'Kesehatan', 'Geografi',
                       'Komunikasi', 'Sejarah dan Filsafat']

index_mapping = {column: index+1 for index, column in enumerate(columns_to_consider)}

# Find the column with the maximum value for each row
max_columns = data[columns_to_consider].idxmax(axis=1)

# Map the column names to their respective index values
data['highest'] = max_columns.map(index_mapping)
```

```python
data
```

```python
index_mapping = {column: index+1 for index, column in enumerate(columns_to_consider)}

# Find the column with the maximum value for each row
max_columns = data_validations[columns_to_consider].idxmax(axis=1)

# Map the column names to their respective index values
data_validations['highest'] = max_columns.map(index_mapping)
data_validations
```

```python
data.to_csv('data_major_with_information.csv', index=False)
data_validations.to_csv('major_data_validation.csv', index=False)
```
