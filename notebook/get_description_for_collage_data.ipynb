```python
import pandas as pd

data_major = pd.read_csv('major_data_for_who.csv')
course_data = pd.read_csv('course_data_cleaned_after_checked.csv')
data_major
```

```python
data_major.info()
data_major.describe()
```

```python
course_data.info()
```

```python
# Gabungkan kolom 'Description', 'for_who', dan 'Mata Kuliah' di data_major
data_major['information'] = data_major['Description'] + " " + data_major['for_who'] + " " + data_major['Mata Kuliah']

# Buat dictionary dari course_data untuk mempermudah penggabungan deskripsi
course_dict = course_data.groupby('NAMA')['Description'].apply(lambda x: ' '.join(x)).to_dict()
```

```python
course_dict
```

```python
def add_course_description(row):
    major_name = row['NAMA']
    course_description = course_dict.get(major_name, "")
    return row['information'] + " " + course_description
```

```python
data_major['information'] = data_major.apply(add_course_description, axis=1)
```

```python
data_major['information']
```

```python
data_major.to_csv('data_major_with_information.csv', index=False)

# Tampilkan beberapa baris hasil untuk verifikasi
print(data_major[['NAMA', 'information']].head())
```
