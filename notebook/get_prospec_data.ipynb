```python

# only takes 5 courses
def split_prospec(propec):
    propec_list = propec.split('\n')
    propec_list = propec_list[:5]
    propec_list = [mk.split('. ', 1)[1] if '. ' in mk else mk for mk in propec_list]
    return propec_list
```

```python
data['Split_prospec'] = data['prospek'].apply(split_prospec)
```
