```python
import pandas as pd
import tensorflow as tf
import numpy as np

data = pd.read_csv('../dataset/major.csv')
data
```

```python
data['Matematika'] = data['Matematika'].astype(float)
data['Sains'] = data['Sains'].astype(float)
data['Fisika'] = data['Fisika'].astype(float)
data['Sosiologi'] = data['Sosiologi'].astype(float)
data['Biologi'] = data['Biologi'].astype(float)
data['Kimia'] = data['Kimia'].astype(float)
data['Teknologi'] = data['Teknologi'].astype(float)
data['Bisnis dan Ekonomi'] = data['Bisnis dan Ekonomi'].astype(float)
data['Seni'] = data['Seni'].astype(float)
data['Sastra dan Linguistik'] = data['Sastra dan Linguistik'].astype(float)
data['Pendidikan'] = data['Pendidikan'].astype(float)
data['Hukum'] = data['Hukum'].astype(float)
data['Lingkungan'] = data['Lingkungan'].astype(float)
data['Kesehatan'] = data['Kesehatan'].astype(float)
data['Geografi'] = data['Geografi'].astype(float)
data['Komunikasi'] = data['Komunikasi'].astype(float)
data['Sejarah dan Filsafat'] = data['Sejarah dan Filsafat'].astype(float)

data.info()
```

```python
data.describe()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dot
from sklearn.preprocessing import normalize
```

```python
features = data[['Matematika', 'Sains', 'Fisika', 'Sosiologi', 'Biologi', 'Kimia', 'Teknologi', 'Bisnis dan Ekonomi', 'Seni',
               'Sastra dan Linguistik', 'Pendidikan', 'Lingkungan', 'Kesehatan', 'Geografi', 'Komunikasi',
               'Sejarah dan Filsafat', 'Hukum']].values
targets = data['NAMA'].values
```

```python
targets
```

```python
# Normalisasi fitur
features_normalized = features
input_features = Input(shape=(features.shape[1],))
```

```python
np.save('features_normalized.npy', features_normalized)
np.save('targets.npy', targets)
```

```python
features_normalized
```

```python
x = Dense(64, activation='relu')(input_features)
x = Dense(32, activation='relu')(x)
output = Dense(features.shape[1], activation='sigmoid')(x)
```

```python
model = Model(inputs=input_features, outputs=output)
model.compile(optimizer='adam', loss='mse')

model.fit(features_normalized, features_normalized, epochs=1000, verbose=0)
```

```python

train_loss = model.evaluate(features_normalized, features_normalized, verbose=0)
print(f'Training Loss (MSE): {train_loss}')


reconstructed = model.predict(features_normalized)
reconstruction_error = np.mean(np.square(features_normalized - reconstructed), axis=1)
print(f'Average Reconstruction Error: {np.mean(reconstruction_error)}')


```

```python
def recommend(input_data, model, targets, top_k=10):
    input_data_normalized = normalize([input_data], axis=1)
    pred = model.predict(input_data_normalized)
    similarities = np.dot(features_normalized, pred.T).flatten()
    top_k_indices = similarities.argsort()[-top_k:][::-1]
    recommended_courses = [targets[i] for i in top_k_indices]

    return recommended_courses

input_data = [0.8, 0.7, 0.9, 0.6, 0.7, 0.8, 0.9, 0.6, 0.8, 0.7, 0.9, 0.6, 0.7, 0.8, 0.9, 0.6, 0.7]
recommendations = recommend(input_data, model, targets)
print(recommendations)
```
